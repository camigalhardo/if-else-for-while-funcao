Crie uma funcao recursiva, potencial(base,expoente), que calcula uma potencia da forma base expoente. Para isso note que a elevado a b = a x a elevado a b-1. Exemplo print(potencial(2,3))

```python
def potencia(base, expoente):
  if expoente ==0:
    return 1
  elif expoente ==1:
    return base

  return base * potencia(base, expoente -1)

print("Insira o valor:")
base = int(input("Base: "))
expoente = int(input("Expoente: "))

potencia(base, expoente)
```
