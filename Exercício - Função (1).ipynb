```python
def alterna(n):
  mult = 1
  for i in range(n+1):
   mult *= -1
   i *= mult
   print(i, end=" ")

alterna(6)
```

```python
def area_triangulo(base , altura):
  area = base * altura / 2
  return area

base1 = 2
altura1 = 5

base2 = 4
altura2 = 5

print(f"area triagulo 1 = {area_triangulo(base1, altura1)} cm*cm")
print(f"area triagulo 2 = {area_triangulo(base2, altura2)} cm*cm")
```

Implemente uma funcao que calcula o perimetro de um retangulo de lados a e b. Note o uso do termo return no exemplo anterior para retornar o perimetro calculado pela funcao:
perimetro_retangulo(a,b)
lembre se de que o perimetro do retangulo e dado por 2xa+2xb:
print(perimetro_retangulo(2,3))

```python
def perimetro_retangulo(a,b):
  perimetro = (2*a) + (2*b)
  return perimetro

print(perimetro_retangulo(10,15))
```
